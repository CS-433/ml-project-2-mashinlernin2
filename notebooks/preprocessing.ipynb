```python
# imports
import sys

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# make sure you are working on an appropiate environment
print(sys.executable)
```

```python
# accessing the videos
path = '../data/pw1_A1_0315_part1.avi'
vid = cv2.VideoCapture(path)
```

```python
amount_of_frames = vid.get(cv2.CAP_PROP_FRAME_COUNT)
amount_of_frames

def get_frame(vid, frame_id):
    vid.set(cv2.CAP_PROP_POS_FRAMES, frame_id)
    ret, frame = vid.read()
    assert(ret==True)
    return frame[:,:,1]
```

```python
frame_id = 10

f1 = get_frame(vid, 10)
f2 = get_frame(vid, 9)
n, m = f1.shape
plt.imshow(np.abs(f1), cmap='gray', vmin=0, vmax=255)

f1s = f1[n-20:,160:200]
f2s = f2[n-20:,160:200]
plt.imshow(cv2.absdiff(f1s-f2s), cmap='gray', vmin=0, vmax=255)
np.abs(f2s - f1s)
```
